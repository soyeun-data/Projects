### 상관관계 분석을 위하여 팀 명을 라벨 인코딩
- 이 때, one-hot-encoding을 사용하지 않은 이유     
1. one-hot-encoding을 이용할 경우 feature의 개수가 늘어나 희소 행렬로 인하여 모델의 정확도가 더 낮아질 우려가 있었음
2. 범주형 변수를 label encoding 할 경우 모델 학습 시, 불필요한 가중치를 줄 수 있는 문제점이 있지만 팀 종류가 총 11팀으로 모델 학습에 영향을 줄 정도는 아니라고 판단      
=> one-hot-encoding 대신 label encoding 활용

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(baseball_data['T_ID'])
baseball_data.loc[:,'T_ID'] = le.transform(baseball_data['T_ID'])
baseball_data.loc[:,'VS_T_ID'] = le.transform(baseball_data['VS_T_ID'])
baseball_data.head()
```

### 데이터 형태가 object인 값(TB_SC)을 라벨 인코딩

```python
le = LabelEncoder()
le.fit(baseball_data['TB_SC'])
baseball_data.loc[:,'TB_SC'] = le.transform(baseball_data['TB_SC'])
baseball_data.head()
```
